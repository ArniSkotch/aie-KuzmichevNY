##### **2. Примеры создания массивов:**
#####   - из обычных списков;
#####   - с использованием функций `zeros`, `ones`, `arange` или аналогичных.

```python
a = np.array([1, 2, 3, 4])  # 1D массив из списка целых чисел
b = np.array([[1.0, 2.0], [3.0, 4.0]])  # 2D массив (матрица) из вложенных списков
Zeros = np.zeros((2, 3))  # матрица 2x3, заполненная нулями
Ones = np.ones(5)  # 1D массив из пяти единиц
ar = np.arange(0, 10, 2)  # последовательность 0,2,4,6,8 в виде 1D массива
```

```python
# Вывод созданных массивов
print("a:", a)
print("b:\n", b)
print("zeros:\n", Zeros)
print("ones:", Ones)
print("ar:", ar)
```

##### **3. Демонстрация атрибутов:**
#####   - вывод формы (shape) и типа данных (dtype) для одного из массивов.

```python
print("--- Атрибуты массива b ---")
print("shape (форма) b:", b.shape) # форма (размерности) массива
print("dtype (тип данных) b:", b.dtype) # тип элементов в массиве
```

##### **4. Примеры индексирования и срезов:**
#####   - получение отдельного элемента;
#####   - извлечение подмассива с помощью среза.

```python
# Отдельные элементы
third_element = a[2]  # индексация идёт с нуля, следовательно третий элемент массива a имеет индекс "2"
print("Третий элемент a (a[2]):", third_element)
b_0_1 = b[0, 1]  # элемент первой строки, второго столбца в матрице b
print("Элемент b[0,1]:", b_0_1)

# Подмассивы, полученные срезом
slice_ar = ar[1:4]  # элементы с индексами 1,2,3 (4 индекс не берётся)
print("\nСрез ar[1:4]:", slice_ar)
col0 = b[:, 0]  # все строки первого столбца
print("Первый столбец b:", col0)
```
